```python
# 1. Ao visualizar a base você percebeu que seria melhor alterar o nome de algumas colunas.
# Mude o ‘ID’ para ‘Identificador’, corrija o nome da coluna que indica a pressão sanguínea, mude a coluna ‘Ocupação’ para ‘Profissão’,
# a coluna ‘Categoria BMI’ está em parte em inglês, substitua para ‘Categoria IMC’.
df_sono.columns
```

```python
# Renomeando as colunas e alterando o DataFrame
df_sono.rename(columns={'ID':'Identificador', 'Pressão sanguíneaaaa':'Pressão sanguínea','Ocupação':'Profissão','Categoria BMI':'Categoria IMC'}, inplace=True)
df_sono.columns
```

```python
# 2. Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões? [‘mean’, np.median, pd.Series.mod]
mode = pd.Series.mode
df_profissao = df_sono.groupby('Profissão')['Duração do sono'].agg(['mean',np.median, mode] )
df_profissao.columns=['Média','Mediana','Moda']
df_profissao
```

```python
# 3. Das pessoas que atuam com engenharia de software qual a porcentagem de obesos?
df_obsesos_eng_dados = df_sono[df_sono['Profissão'] == 'Eng. de Software']
df_obsesos_eng_dados
```

```python
df_percentual_obeso=np.where(df_obsesos_eng_dados['Categoria IMC'] == 'Obesidade',1,0)
percentual_obeso = df_percentual_obeso.sum()/len(df_percentual_obeso)*100
print("O dataset tem ", percentual_obeso, "% de obesos dentre os profissionais de engenheria de software.")
```

```python
# 4.De acordo com os dados, advogar ou ser representante de vendas faz você dormir menos? (Use o método ‘isin’, considere a média)
df_dormir_menos=df_sono[df_sono['Profissão'].isin(['Advogado(a)','Representante de Vendas'])].groupby('Profissão',as_index=False)['Duração do sono'].mean().round(2).sort_values('Duração do sono')
df_dormir_menos
```


De acordo com os dados, ser Represente de Vendas faz você dormir menos.

```python
# 5. Entre quem fez enfermagem e quem fez medicina, quem tem menos horas de sono? (Use o método ‘isin’, considere a média)
df_dormir_menos=df_sono[df_sono['Profissão'].isin(['Enfermeiro(a)','Médico(a)'])].groupby('Profissão',as_index=False)['Duração do sono'].mean().round(2).sort_values('Duração do sono')
df_dormir_menos
```

De acordo com os dados, há uma diferença sutil entre a média de horas de sono do profissional Enfermeiro(a) e Médico(a). No entanto, os dados indicam, que ser Médico(a) faz você dormir menos.

```python
# 8. Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)
# PAM é o cálculo da Pressão Arterial Média, onde PAM = (PAS + 2 X PAD)  /3
# Onde, PAS (Pressão Arterial Sistólica) e PAD (Pressão Arterial Diastólica)

# cópia do dataframe principal
df_pressao = df_sono.copy()
df_pressao[['PAS','PAD']] = df_pressao['Pressão sanguínea'].str.split('/',expand=True)
df_pressao['PAS']=df_pressao['PAS'].astype(int)
df_pressao['PAD']=df_pressao['PAD'].astype(int)
df_pressao['MediaPA']=round((df_pressao['PAS']+2*df_pressao['PAD'])/3,2)
df_pressao
```

```python
df_pressao.groupby('Gênero',as_index=False)['MediaPA'].mean().round(2).sort_values('MediaPA',ascending=False)
```

De acordo com o conjunto de dados, entre homens e mulheres quem possui a maior pressão sanguínea média são as mulheres.

```python
# 9. É predominante entre os participantes dormir 8 horas por dia (considere usar Moda como medida)?
mode = pd.Series.mode
df_sono['Duração do sono'].mode()
```

```python
df_sono['Duração do sono'].median()
```

```python
# 10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)
frequencia_maior_70 = df_sono[df_sono['Frequência cardíaca'] > 70]['Passos diários'].mean().round(2)
frequencia_maior_70
```

```python
frequencia_menorigual_70 = df_sono[df_sono['Frequência cardíaca'] <= 70]['Passos diários'].mean().round(2)
frequencia_menorigual_70
```

```python
frequencia_maior_70 > frequencia_menorigual_70
```

Dentre os dados no dataset é possível observar que a questão é falsa, tendo em vista que as pessoas com frequências cardíacas menor ou igual a 70 dão mais passos diários que aquelas com frequência cardíaca maior acima de 70.
